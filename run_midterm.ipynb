```python
with open("result.txt", "r") as f:
  lines = f.readlines()

count = 1
ok = 0
for line in lines:
  num = int(line)

  if 0 < count < 11:
    if num == 0:
      ok += 1
  elif 10 < count < 21:
    if num == 1:
      ok += 1
  elif 20 < count < 31:
    if num == 2:
      ok += 1
  elif 30 < count < 41:
    if num == 3:
      ok += 1
  elif 40 < count < 51:
    if num == 4:
      ok += 1
  elif 50 < count < 61:
    if num == 5:
      ok += 1
  elif 60 < count < 71:
    if num == 6:
      ok += 1
  elif 70 < count < 81:
    if num == 7:
      ok += 1
  elif 80 < count < 91:
    if num == 8:
      ok += 1
  elif 90 < count < 101:
    if num == 9:
      ok += 1

  count += 1

print("Test accuracy: " + str(ok) + "%")
```
